# tessvectors convenience functions
The full available vectors and diagnostic files can be found at:

https://heasarc.gsfc.nasa.gov/docs/tess/data/TESSVectors/

The `tessvectors` package also contains a few convenience functions in a lightweight API that may help you interact with these data. 

## getvector - to read and download tessvector files
However, this can be awkward to access programatically through python.  To simply this, the `tessvectors.getvector` function will help you read, and download, a vector for a given observing cadence, sector, and camera.  These can be specified directly, or the path to a fits file, or a fits hdulist from an opened fits file can be passed.  In the event of a fits path or file, a SPOC like processing is assumed to access the appropriate header keywords. To download a file for later use, pass the optional parameter `download=True`. 

```python
import tessvectors
vector = tessvectors.getvector(("FFI", 1, 4))
vector
```

```python
# Next, to demonstrate the fits api, lets get a tpf via lightkurve
import lightkurve as lk
tpf = lk.search_targetpixelfile(
        "Pi Men C", sector=1, exptime="short", author="SPOC"
    ).download()
```

```python
print(tpf.path)
vector = tessvectors.getvector(tpf.path)
vector
```

```python
print(tpf.hdu)
vector = tessvectors.getvector(tpf.hdu)
vector
```

### getvectors - download
we can download a file for later use by passing the `download = True` optional parameter, e.g.:

```python
vector = tessvectors.getvector(tpf.path, download=True)
vector
```

## tessvectors getFFI - get the URI of an FFI file for a specified Cadence/Sector/Camera/CCD

Sometimes when looking at tpfs from 20s or 120s cadence data, you may wish to view a FFI taken at approximately the same time that corresponds to a particular frame of the high cadence data (albeit binned with additional data in time from nearby frames of the 120s/20s data).  This might be to examine the current effects of systematics such as scattered background light or nearby variable stars.  

TESS FFI's are available via the MAST bulk download page at https://archive.stsci.edu/tess/bulk_downloads/bulk_downloads_ffi-tp-lc-dv.html. Finding the particular FFI that you are looking for can be tricky, however. TPFs created with TESSCut have the FFI assosciated with each frame available as part of the header file.  For TPF from 120s or 20s data, however, it can be more difficult to figure out which FFI has data assosciated with a particular frame from the higher-cadence product.  The FFI assosciated with each frame of a TESSCut TPF has been stored in the `FFI` cadence `tessvector` file.  We can then crossmatch between the midtime of our fram of interest in a 20s or 120s tpf and the FFI's.  

To make this process simpler, `tessvectors` has the convenient `getffi` function that will use the above information to determine the appropriate FFI and its URI.  This can then be downloaded through entering it into your web-browser or through a command-line tool such as wget.  

getffi takes a tuple that corresponds to (midtime, cadence, sector, camera, ccd)

```python
ffi_loc = tessvectors.getffi((1337.370729227238,1,4,1))
ffi_loc
```
